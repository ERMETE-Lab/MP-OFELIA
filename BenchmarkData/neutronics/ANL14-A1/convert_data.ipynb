```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
import pandas as pd
```

# Power

```python
# Load benchmark data
df = pd.read_csv('./csv_data/power_0.csv')
power_data = dict()
power_data['t'] = df.to_numpy()[:, 0]
power_data['Power'] = df.to_numpy()[:, 1]
```

```python
df = pd.read_csv('./csv_data/power_1.csv')
power_data['t'] = np.hstack([power_data['t'], df.to_numpy()[:, 0]])
power_data['Power'] = np.hstack([power_data['Power'], df.to_numpy()[:, 1]])
```

```python
df = pd.read_csv('./csv_data/power_2.csv')
power_data['t'] = np.hstack([power_data['t'], df.to_numpy()[:, 0]])
power_data['Power'] = np.hstack([power_data['Power'], df.to_numpy()[:, 1]])
```

```python
df = pd.read_csv('./csv_data/power_3.csv')
power_data['t'] = np.hstack([power_data['t'], df.to_numpy()[:, 0]])
power_data['Power'] = np.hstack([power_data['Power'], df.to_numpy()[:, 1]])
```

```python
df = pd.read_csv('./csv_data/power_4.csv')
power_data['t'] = np.hstack([power_data['t'], df.to_numpy()[:, 0]])
power_data['Power'] = np.hstack([power_data['Power'], df.to_numpy()[:, 1]])
```

```python
plt.semilogy(power_data['t'], power_data['Power'], 'o-')
plt.grid()
```

```python
with open('LRA_power.data', 'wb') as handle:
    pickle.dump(power_data, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
